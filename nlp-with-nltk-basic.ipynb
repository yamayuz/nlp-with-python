nltk.text.Textに含まれるメソッドたち

concordance(): 引数の単語が含まれる文を探す<br>
count(): 文章中に単語が出てきた回数を数える（listのcount()がtokensに適用される）<br>
common_context(): 複数の単語に共通する表現を見つける<br>
dispersion_plot(): 時系列に沿って単語の出現をプロットする<br>
index(): 引数の単語が初めて文章中に出現するインデックスを取得する（listのindex()がtokensに適用される）<br>
similar(): 似たような分布の単語を見つける<br>
babelize_shell(): 2言語間の翻訳を平衡状態（Equilibrium）になるまで繰り返す。廃止<br>
